```python
lat_long = pd.read_html("https://www.latlong.net/category/states-236-14.html")[0]
lat_long
```

```python
# new data frame with split value columns
new = lat_long["Place Name"].str.split(",", n = 1, expand = True)
  
# making separate first name column from new data frame
lat_long["States"]= new[0]
  
# making separate last name column from new data frame
lat_long["Country"]= new[1]
  
# Dropping old Name columns
lat_long.drop(columns =["Place Name"], inplace = True)
lat_long
```

```python
state_cord = lat_long[['States','Latitude','Longitude']].copy()
state_cord.sort_values(by=['States']).reset_index(drop=True, inplace=True)
```

```python
state_cord
```

```python
state_cord.to_csv("../Output/state_cord.csv")
```

```python
# Engine connection
rds_connection_string = "postgres:postgres@localhost:5432/Real_Estate"
engine = create_engine(f'postgresql://{rds_connection_string}')

# Verifying that connection is made
engine.table_names()

# Loading data into Postgres Database 
state_cord.to_sql(name='State_Cords', con=engine, if_exists='append', index=False)
```
